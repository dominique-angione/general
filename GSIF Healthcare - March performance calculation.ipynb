```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
# Active name performance
```

```python
tickers = ['BSX', 'WCG', 'SYNH']
mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader(t, data_source='yahoo', start='2019-3-1', end='2019-4-1')['Adj Close']
```

```python
mydata.iloc[0]
```

```python
(mydata / mydata.iloc[0] * 100).plot(figsize = (10, 6))
```

```python
one_month_returns = mydata.iloc[-1] / mydata.iloc[0] - 1 
```

```python
one_month_returns
```

```python
weights = np.array([.3, .2, .5])
```

```python
np.dot(one_month_returns, weights)
```

```python
march_performance = str(round(np.dot(one_month_returns, weights), 5) * 100) + ' %'
print (march_performance)
```
